```python
input = []
with open('input.txt') as f:
    for line in f:
        input.append(list(map(int, line.strip().split())))

```

```python
def solution(input):

    diff = input

    for i in range(len(input)):
        for j in range(len(input[i])-1):
            diff[i][j] = input[i][j+1] - input[i][j]
        diff[i].pop()

    Pass = []
    for i in range(len(diff)):
        if all((x > 0 and x < 4) for x in diff[i]) or all((x < 0 and x > -4) for x in diff[i]) :
            Pass.append(i)
    return Pass
```

```python
def singleSolution(input):
    diff = input
    for i in range(len(input)-1):
        diff[i] = input[i+1] - input[i]
    diff.pop()
    if all((ele > 0 and ele < 4) for ele in diff) or all((ele < 0 and ele > -4) for ele in diff) :
        return True
    return False
```

```python
Pass = solution(input)
Fail = [i for i in range(len(input)) if i not in Pass]
```

```python
failInput = [input[row][:] for row in Fail]


for i in failInput:
    temp = i[:]

    for j in range(len(temp)-1,-1,-1):
        print(temp)
        temp.pop(j)
        if singleSolution(temp) == True:
            print(True)
            Pass.append(i) 
            break
            
        temp = i[:]

        
print(len(Pass))
```
